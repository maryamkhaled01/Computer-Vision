```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
# Load image
img = cv2.imread('/content/image.png')

# Display image
cv2_imshow(img)
```

```python
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
# Apply median filter for noise reduction
gray_blur = cv2.medianBlur(gray, 7)
cv2_imshow(gray_blur)  # Display noise-reduced image
```

```python
# Edge Detection with Laplacian filter
edges = cv2.Laplacian(gray_blur, cv2.CV_8U, ksize=5) # Higher ksize gives more smoothing
cv2_imshow(edges) # Display edge detection result
```

```python
# Edge thresholding
_, sketch = cv2.threshold(edges, 140, 255, cv2.THRESH_BINARY_INV)
cv2_imshow(sketch) # Display Sketch
```

```python
# Create a color painting using bilateral filter
color = img
for i in range (3):  # Apply bilateral filter 3 times
  color = cv2.bilateralFilter(color, 9, 10, 50)
  cv2_imshow(color)  # Display color painting
```

```python
# Combine sketch and color painting for final cartoon effect
cartoon = cv2.bitwise_and(color, color, mask=sketch)
cv2_imshow(cartoon)  # Display cartoon effect
```
